# Exploring Star Wars Fans: Uncovering Insights

In this notebook, I have cleaned and explored the dataset from a FiveThirtyEight survey on Star Wars fans' perceptions. The dataset covers demographics like gender, age, income, and education, along with responses on Star Wars movie viewership and opinions. Our goal is to analyz this data, we aim to uncover insights into Star Wars fans' preferences and perceptions, contributing to a deeper understanding of the franchise's cultural significance.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = star_wars.copy()
# droping these column as the name of characters are not mentioned
df.drop(columns=['Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.',
 'Unnamed: 16',
 'Unnamed: 17',
 'Unnamed: 18',
 'Unnamed: 19',
 'Unnamed: 20',
 'Unnamed: 21',
 'Unnamed: 22',
 'Unnamed: 23',
 'Unnamed: 24',
 'Unnamed: 25',
 'Unnamed: 26',
 'Unnamed: 27',
 'Unnamed: 28'],inplace=True)
```

```python

# Dictionary mapping original column names to new shorter names
column_name_mapping = {
    'RespondentID': 'ID',
    'Have you seen any of the 6 films in the Star Wars franchise?': 'SeenAnySW',
    'Do you consider yourself to be a fan of the Star Wars film franchise?': 'SWFan',
    'Which of the following Star Wars films have you seen? Please select all that apply.': 'SWFilm1Seen',
    'Unnamed: 4': 'SWFilm2Seen',
    'Unnamed: 5': 'SWFilm3Seen',
    'Unnamed: 6': 'SWFilm4Seen',
    'Unnamed: 7': 'SWFilm5Seen',
    'Unnamed: 8': 'SWFilm6Seen',
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.': 'SWFilm1Rank',
    'Unnamed: 10': 'SWFilm2Rank',
    'Unnamed: 11': 'SWFilm3Rank',
    'Unnamed: 12': 'SWFilm4Rank',
    'Unnamed: 13': 'SWFilm5Rank',
    'Unnamed: 14': 'SWFilm6Rank',
    'Which character shot first?': 'FirstShot',
    'Are you familiar with the Expanded Universe?': 'KnowEU',
    'Do you consider yourself to be a fan of the Expanded Universe?': 'EUFan',
    'Do you consider yourself to be a fan of the Star Trek franchise?': 'STFan',
    'Gender': 'Gender',
    'Age': 'Age',
    'Household Income': 'Income',
    'Education': 'Education',
    'Location (Census Region)': 'Location'
}

df.rename(columns=column_name_mapping, inplace=True)
```

```python
df.head()
```

```python
df[['SWFilm1Seen', 'SWFilm2Seen', 'SWFilm3Seen', 'SWFilm4Seen', 'SWFilm5Seen', 'SWFilm6Seen']] = (
    np.where(df[['SWFilm1Seen', 'SWFilm2Seen', 'SWFilm3Seen', 'SWFilm4Seen', 'SWFilm5Seen', 'SWFilm6Seen']].notnull(), True, False)
)
```

```python
df[['SWFilm1Seen','SWFilm2Seen','SWFilm3Seen','SWFilm4Seen','SWFilm5Seen','SWFilm6Seen']]
```

#### Converting *SeenAnySW* and *SWFan* to boolean series

```python
df[['SeenAnySW','SWFan']]=df[['SeenAnySW','SWFan']].apply(lambda x: x=='Yes') 
```

#### Converting ranking columns to float type

```python
df.iloc[:,9:15] = df.iloc[:,9:15].astype(float)
df.iloc[:,9:15]
```

### Strategy Explanation

The values in these columns are ranks, where **1** is the highest rank and **6** is the lowest. To make these values more intuitive, we can convert the ranks into scores. This can be achieved by subtracting each rank from **7**, which will transform the ranking system into a scoring system where a higher rank results in a higher score.

### Conversion Formula

**Score** = **7** - **Rank**

### Example

- Rank 1 becomes: \( 7 - 1 = 6 \)
- Rank 2 becomes: \( 7 - 2 = 5 \)
- Rank 3 becomes: \( 7 - 3 = 4 \)
- Rank 4 becomes: \( 7 - 4 = 3 \)
- Rank 5 becomes: \( 7 - 5 = 2 \)
- Rank 6 becomes: \( 7 - 6 = 1 \)

By converting ranks to scores, we can make the data more intuitive and easier to interpret.


```python
score = 7-df.iloc[:,9:15]
```

```python
# df.iloc[:,9:15].mean(axis = 0).plot(kind='bar')
score.mean(axis = 0).plot(kind='bar')
plt.xticks(rotation=45)
plt.title("Starwars Ranking")
plt.show()
```

`Star Wars: Episode V - The Empire Strikes Back` is the most liked film, followed by `Star Wars: Episode VI Return of the Jedi`. `Star Wars: Episode III - Revenge of the Sith`` ranks last`.

```python
males = df['Gender']=='Male'
females = df['Gender']=='Female'
```

```python
plt.figure(figsize=(8, 4))

# Male viewership subplot
plt.subplot(1, 2, 1)  # 2 rows, 1 column, subplot 1
score[males].sum().plot(kind='bar')
plt.title("Ranking by Males")

# Female viewership subplot
plt.subplot(1, 2, 2)  # 2 rows, 1 column, subplot 2
score[females].sum().plot(kind='bar')
plt.title("Ranking by Females")

plt.tight_layout()
plt.show()
```

##### The ranking trend among fans remains consistent across both genders.

## Analyzing Viewership

```python
df.iloc[:,3:9].sum().plot(kind='bar')
plt.title("Star Wars Movie Viewership")
plt.xticks(rotation=45)
plt.show()
```

The bar graph visualizes the viewership counts for each Star Wars movie based on the provided data. It reveals the following insights:
- `SWFilm5Seen` and `SWFilm6Seen` are the most popular Star Wars movies among the sampled audience.
- `SWFilm1Seen` and `SWFilm4Seen` have lower popularity.
- `SWFilm3Seen` has the least popularity among the sampled audience.


```python
plt.figure(figsize=(8, 4))

# Male viewership subplot
plt.subplot(1, 2, 1)  # 2 rows, 1 column, subplot 1
df[males].iloc[:, 3:9].sum().plot(kind='bar')
plt.title("Male Viewership")

# Female viewership subplot
plt.subplot(1, 2, 2)  # 2 rows, 1 column, subplot 2
df[females].iloc[:, 3:9].sum().plot(kind='bar')
plt.title("Female Viewership")

plt.tight_layout()
plt.show()
```
